```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('final_data.csv')
#df.head()
```

```python
df_death=df[df['isAlive']==0]
houses=df_death['house'].dropna().tolist()
#print(houses)
```

```python
def draw_cloud(word_list,max_=20):
    from wordcloud import WordCloud
    import matplotlib.pyplot as plt
    %matplotlib inline
    from random import shuffle
    shuffle(word_list)
    text = ' '.join(word_list)
    text=text.replace('House','')
    wordcloud = WordCloud(background_color='white',max_words=max_).generate(text)
    plt.figure(figsize=[20,10])
    plt.axis('off')
    plt.imshow(wordcloud)
    return
```

```python
draw_cloud(houses,max_=50)
```
